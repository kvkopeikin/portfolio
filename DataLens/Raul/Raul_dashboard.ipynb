# Дашборд отображающий динамику и структуру трудоустройства

Для благотворительного фонда Рауль необходимо подготовить дашборд отображающий динамику и структуру трудоустройства кандидатов.

Требования к дашборду:

- Параметры динамики:
    - помесячно,
    - поквартально,
    - ежегодно.
    
- Параметры структуры:
    - принадлежность благополучателя к одному из сегментов,
    - ответственная НКО,
    - регион,
    - пол благополучателя,
    - возраст,
    - тип трудоустройства (временное, постоянное с сопровождением, постоянное без сопровождения),
    - тип работодателя (внутренняя защищённая площадка, внешняя защищённая площадка, открытый рынок).
    
- Параметры трудоустройства: от 1 дня, от 1 месяца

Импортирую необходимые библиотеки:

```python
import pandas as pd
from io import BytesIO
import requests
import datetime
```

Прочитаю данные из файла на гугл диске:

```python
spreadsheet_id = ''
file_name = 'https://docs.google.com/spreadsheets/d/{}/export?format=csv'.format(spreadsheet_id)
r = requests.get(file_name)
data = pd.read_csv(BytesIO(r.content))
```

Посмотрю общую информацию о массиве:

```python
data.head(10)
```

```python
data.info()
```

**Типы данных** в некоторых столбцах не соответствуют содержимому, преобразую их:

```python
columns = ['инвалидность','сиротский_опыт','адаптированная_образовательная_программа','1+_месяц_работы']

for column in columns:
    data[column] = data[column].replace(['ИСТИНА','ЛОЖЬ'],[True,False])
```

```python
data[['дата_добавления', 'начало_трудоустройства','завершение_трудоустройства']] = \
data[['дата_добавления', 'начало_трудоустройства','завершение_трудоустройства']].apply(pd.to_datetime,
                                                                                       format='%d.%m.%Y %H:%M:%S')
```

```python
data.info()
```

Проверю данные на **дубликаты**:

```python
print('Дубликатов в таблице data:',data.duplicated().sum())
```

Проверю массив на **пропуски**:

```python
data.isna().sum()
```

Очевидно что трудоустройство может быть не завершено - эти пропуски заполнять нет смысла, остальные пропуски заполнить не представляется возможным, однако необходимо известить заказчика о них. Посмотрю детальнее на строки с пропусками:

```python
display(data.loc[data['дата_добавления'].isna() == True])
display(data.loc[data['тип_площадки'].isna() == True])
display(data.loc[data['тип_трудоустройства'].isna() == True])
display(data.loc[data['тип_опыта_работы'].isna() == True])
```

В итоге имеем неполные данные для кандидата с id = 213 - отсутствует дата добавления и для кандидата с id = 197 отстутствуют данные о типе площадки, типе трудоустройства и типе опыта работы.

Оценю общую адекватность данных, посмотрю максимальные и минимальные значения, характер распределения данных:

```python
data.describe()
```

Аномальных значений не обнаружено, для построения имеем данные по трудоустройсту с ноября 2022 по август 2023 года,возраст кандидатов от 14 до 39 лет.  

Количество уникальных кандидатов:

```python
len(data.id_кандидата.unique())
```

Помимо этого имеем неполные данные для кандидата с id = 213 - отсутствует дата добавления и для кандидата с id = 197 отстутствуют данные о типе площадки, типе трудоустройства и типе опыта работы - эти данные исказят результаты дашборда, поскольку данные кандидаты будут исключены при фильтрации по соответствующим признакам.

Аномальных значений не обнаружено, для построения дашборда имеем данные по трудоустройсту с ноября 2022 по август 2023 года,возраст кандидатов от 14 до 39 лет.
